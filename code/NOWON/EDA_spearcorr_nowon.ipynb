# Spearcorr를 이용하여 실제 상관관계 파악

```python
import pandas as pd # 데이터 프레임 처리
import numpy as np # 벡터 계산
import matplotlib.pyplot as plt # 데이터 시각화
import seaborn as sns
# 사이킷런 패키지 활용
from scipy.stats import spearmanr
```

```python
total_20 = pd.read_csv('../data/innovation/Merge_data/df_age20_nowon.csv') # 20대 노원구 데이터
total_30 = pd.read_csv('../data/innovation/Merge_data/df_age30_nowon.csv') # 30대 노원구 데이터
total_40 = pd.read_csv('../data/innovation/Merge_data/df_age40_nowon.csv') # 40대 노원구 데이터
total_50 = pd.read_csv('../data/innovation/Merge_data/df_age50_nowon.csv') # 50대 노원구 데이터
```

```python
total_20.head()
```

```python
# 시계열 데이터로 만들기 위해서 기존의 인덱스를 2018.04.01 ~ 2019.03.31로 변경
dateIdx = pd.date_range('2018-04-01', '2019-03-31')

total_20.index = dateIdx
total_30.index = dateIdx
total_40.index = dateIdx
total_50.index = dateIdx
```

# 상관계수구하기

```python
corrs_CONTENT_20 = []
corrs_pm10_20 = []
corrs_pm25_20 = []
corrs_temp_20 = []
corrs_humi_20 = []
corrs_rain_20 = []

for c in list(total_20.columns)[:-6] :
    corrs_CONTENT_20.append(spearmanr(total_20[["CONTENT"]], total_20[[c]])[0])
    corrs_pm10_20.append(spearmanr(total_20[["pm10"]], total_20[[c]])[0])
    corrs_pm25_20.append(spearmanr(total_20[["pm25"]], total_20[[c]])[0])
    corrs_temp_20.append(spearmanr(total_20[["temp"]], total_20[[c]])[0])
    corrs_humi_20.append(spearmanr(total_20[["humi"]], total_20[[c]])[0])
    corrs_rain_20.append(spearmanr(total_20[["rain"]], total_20[[c]])[0])
    
corrs_dict_20 = {"CONTENT":corrs_CONTENT_20, "pm10":corrs_pm10_20, "pm25":corrs_pm25_20, "temp":corrs_temp_20, "humi":corrs_humi_20, "rain":corrs_rain_20}
```

```python
corrs_CONTENT_30 = []
corrs_pm10_30 = []
corrs_pm25_30 = []
corrs_temp_30 = []
corrs_humi_30 = []
corrs_rain_30 = []

for c in list(total_30.columns)[:-6] :
    corrs_CONTENT_30.append(spearmanr(total_30[["CONTENT"]], total_30[[c]])[0])
    corrs_pm10_30.append(spearmanr(total_30[["pm10"]], total_30[[c]])[0])
    corrs_pm25_30.append(spearmanr(total_30[["pm25"]], total_30[[c]])[0])
    corrs_temp_30.append(spearmanr(total_30[["temp"]], total_30[[c]])[0])
    corrs_humi_30.append(spearmanr(total_30[["humi"]], total_30[[c]])[0])
    corrs_rain_30.append(spearmanr(total_30[["rain"]], total_30[[c]])[0])
    
corrs_dict_30 = {"CONTENT":corrs_CONTENT_30, "pm10":corrs_pm10_30, "pm25":corrs_pm25_30, "temp":corrs_temp_30, "humi":corrs_humi_30, "rain":corrs_rain_30}
```

```python
corrs_CONTENT_40 = []
corrs_pm10_40 = []
corrs_pm25_40 = []
corrs_temp_40 = []
corrs_humi_40 = []
corrs_rain_40 = []

for c in list(total_40.columns)[:-6] :
    corrs_CONTENT_40.append(spearmanr(total_40[["CONTENT"]], total_40[[c]])[0])
    corrs_pm10_40.append(spearmanr(total_40[["pm10"]], total_40[[c]])[0])
    corrs_pm25_40.append(spearmanr(total_40[["pm25"]], total_40[[c]])[0])
    corrs_temp_40.append(spearmanr(total_40[["temp"]], total_40[[c]])[0])
    corrs_humi_40.append(spearmanr(total_40[["humi"]], total_40[[c]])[0])
    corrs_rain_40.append(spearmanr(total_40[["rain"]], total_40[[c]])[0])
    
corrs_dict_40 = {"CONTENT":corrs_CONTENT_40, "pm10":corrs_pm10_40, "pm25":corrs_pm25_40, "temp":corrs_temp_40, "humi":corrs_humi_40, "rain":corrs_rain_40}
```

```python
corrs_CONTENT_50 = []
corrs_pm10_50 = []
corrs_pm25_50 = []
corrs_temp_50 = []
corrs_humi_50 = []
corrs_rain_50 = []

for c in list(total_50.columns)[:-6] :
    corrs_CONTENT_50.append(spearmanr(total_50[["CONTENT"]], total_50[[c]])[0])
    corrs_pm10_50.append(spearmanr(total_50[["pm10"]], total_50[[c]])[0])
    corrs_pm25_50.append(spearmanr(total_50[["pm25"]], total_50[[c]])[0])
    corrs_temp_50.append(spearmanr(total_50[["temp"]], total_50[[c]])[0])
    corrs_humi_50.append(spearmanr(total_50[["humi"]], total_50[[c]])[0])
    corrs_rain_50.append(spearmanr(total_50[["rain"]], total_50[[c]])[0])
    
corrs_dict_50 = {"CONTENT":corrs_CONTENT_50, "pm10":corrs_pm10_50, "pm25":corrs_pm25_50, "temp":corrs_temp_50, "humi":corrs_humi_50, "rain":corrs_rain_50}
```

```python
sp_df_20 = pd.DataFrame(corrs_dict_20, columns = total_20.columns[-6:], index = total_20.columns[:-6])
sp_df_30 = pd.DataFrame(corrs_dict_30, columns = total_30.columns[-6:], index = total_30.columns[:-6])
sp_df_40 = pd.DataFrame(corrs_dict_40, columns = total_40.columns[-6:], index = total_40.columns[:-6])
sp_df_50 = pd.DataFrame(corrs_dict_50, columns = total_50.columns[-6:], index = total_50.columns[:-6])
```

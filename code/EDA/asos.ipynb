# 미세먼지 측정소별 차이 경향보기

```python
df_asos_no15 = df_asos_no15.groupby('DATE').mean()
df_asos_no15 = df_asos_no15.reset_index()

df_asos_no14 = df_asos_no14.groupby('DATE').mean()
df_asos_no14 = df_asos_no14.reset_index()

df_asos_jong13 = df_asos_jong13.groupby('DATE').mean()
df_asos_jong13 = df_asos_jong13.reset_index()

df_asos_jong12 = df_asos_jong12.groupby('DATE').mean()
df_asos_jong12 = df_asos_jong12.reset_index()
```

```python
df_asos_no15['GU_NM']=['노원구']*len(df_asos_no15)
df_asos_no14['GU_NM']=['노원구']*len(df_asos_no14)
df_asos_jong13['GU_NM']=['종로구']*len(df_asos_jong13)
df_asos_jong12['GU_NM']=['종로구']*len(df_asos_jong12)
```

```python
df_asos1 = pd.concat([df_asos_no15, df_asos_jong13]).groupby(['DATE', 'GU_NM']).sum().reset_index()  # 어차피 pk가 날짜, 구_이름이므로 SUM해도 괜찮다.
df_asos2 = pd.concat([df_asos_no14, df_asos_jong12]).groupby(['DATE', 'GU_NM']).sum().reset_index()
```

# df_asos1로 사용하기

```python
df_asos_nowon = df_asos1.loc[df_asos1.GU_NM == "노원구"]
df_asos_jongro = df_asos1.loc[df_asos1.GU_NM == "종로구"]
```

```python
dateidx = pd.date_range('2018-04-01', '2019-03-31')
df_asos_nowon.index = dateidx
df_asos_jongro.index = dateidx
```

```python
df_asos_nowon.drop(["GU_NM", "DATE"], axis = 1, inplace = True)
df_asos_jongro.drop(["GU_NM", "DATE"], axis = 1, inplace = True)
```

```python
df_total_nowon = df_asos_nowon.merge(df_news, on=df_news.index).set_index('key_0')
df_total_jongro = df_asos_jongro.merge(df_news, on=df_news.index).set_index('key_0')
```

```python
df_total_nowon = df_total_nowon.merge(df_rain, on=df_total_nowon.index).set_index('key_0')
df_total_jongro = df_total_jongro.merge(df_rain, on=df_total_jongro.index).set_index('key_0')
```

```python
df_total_nowon.index=dateidx
df_total_jongro.index=dateidx
```
